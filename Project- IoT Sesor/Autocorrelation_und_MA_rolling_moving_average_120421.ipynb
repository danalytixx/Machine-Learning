## plot the temp data per date

```python
from pandas import read_csv
#from matplotlib import pyplot
import matplotlib.pyplot as plt
import plotly.express as px

plt.figure(figsize=(10, 10))

location="https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"

series = read_csv(location, header=0, index_col=0)
#series.plot()




fig = px.line(x=series.index, y=series['Temp'], title='temp per date')
fig.show()

plt.show()
```

## calc rolling window of temperature
https://towardsdatascience.com/making-a-trade-call-using-simple-moving-average-sma-crossover-strategy-python-implementation-29963326da7a 


```python
series['SMA_Temp']=series['Temp'].rolling(window = 20, min_periods = 1).mean()

series
```

```python
import matplotlib.pyplot as plt

#fig = px.line(x=series.index, y=series['Temp'], title='temp per date')
fig=px.line(x=series.index, y=series['SMA_Temp'], title='rolling MA temp per date')
fig.show()

#plt.show()
```

```python
import pandas as pd

df2 = pd.DataFrame([[0]*series.shape[1]],columns=series.columns)
df3 = series.append(df2, ignore_index=True)
df3
```

## plot ACF of the signal

```python

from pandas import read_csv
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf
series = read_csv(location, header=0, index_col=0)


plot_acf(series, lags=3649)
pyplot.show()
```
