## plot the temp data per date

```python
from pandas import read_csv
#from matplotlib import pyplot
import matplotlib.pyplot as plt
import plotly.express as px

plt.figure(figsize=(10, 10))

location="https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"

series = read_csv(location, header=0, index_col=0)
#series.plot()

series['index'] = range(1, len(series) + 1)

#series.values.reshape(-1,1)



series.shape
series


fig = px.line(x=series['index'], y=series['Temp'], title='temp per date')
fig.show()

plt.show()
```

```python
ser_res=series.values.reshape(-1,1)
ser_res.shape
```

```python
index_res=series['index'].values.reshape(-1,1)
```

## calc the std deviations and varinace

```python
import numpy as np

# calc the standard deviation

n = len(series['Temp'])
# Mean of the data
mean = sum(series['Temp']) / n
# Square deviations
deviations = [(x - mean) ** 2 for x in series['Temp']]
# Variance
variance = sum(deviations) / n

deviations

variance

#cal the std deviation of the signal
std_deviation=np.std(series['Temp'])
```

## plot SVR with the original signal and std and moving average

```python
  from sklearn.linear_model import LinearRegression # for building a linear regression model
from sklearn.svm import SVR # for building support vector regression model
import plotly.graph_objects as go # for data visualization
import plotly.express as px # for data visualization


#calc the moving average
series['SMA_Temp']=series['Temp'].rolling(window = 20, min_periods = 1).mean()

#define the epsilon area which is the std deviation
epsilon=2*std_deviation




# train the model

model = SVR(kernel='rbf', C=1, epsilon=epsilon)
model.fit(index_res,series['Temp'])

y_hat=model.predict(index_res)

fig = px.line(x=series['index'], y=series['Temp'], title='temperature per date')

fig.add_traces(go.Scatter(x=series['index'], y=y_hat, name='pred temp per date', line=dict(color='green')))
fig.add_traces(go.Scatter(x=series['index'], y=series['SMA_Temp'], name='moving average', line=dict(color='black')))
fig.add_traces(go.Scatter(x=series['index'], y=y_hat+epsilon, name='+ epsilon',line=dict(color='red', dash='dot'))) 
fig.add_traces(go.Scatter(x=series['index'], y=y_hat-epsilon, name='- epsilon',line=dict(color='red', dash='dot'))) #,line=dict(color='red', dash='dot'))
fig.show()
```
